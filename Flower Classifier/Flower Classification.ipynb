```python
import os
import matplotlib.pyplot as plt
import numpy as np
from skimage.io import imread
from skimage.transform import resize
```

```python
#Preprocessing
#1.Resize the image
#2.Flatten the image

target=[]
images=[]
flat_data=[]


Data_dir=r"C:\Users\Dell\Downloads\Image Classification\images"
Categories=['Jasmine','Rose','Sunflower']

for cat in Categories:
    class_num=Categories.index(cat)#Label Encoding
    path=os.path.join(Data_dir,cat)#Create path to use all images
    for img in os.listdir(path):
        img_array=imread(os.path.join(path,img))
        #plt.imshow(img_array)
        img_resized=resize(img_array,(150,150,3))#Normalize the image in 0 to 1
        flat_data.append(img_resized.flatten())
        images.append(img_resized)
        target.append(class_num)
        
flat_data=np.array(flat_data)
target=np.array(target)
images=np.array(images)
    
```

```python
np.unique(target,return_counts=True)
```

```python
# Spliting Data in train and test

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(flat_data,target,test_size=0.3,
                                              random_state=110)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm

param_grid=[
    {"C":[1,100,10,1000],'kernel':["linear"]},
    {"C":[1,100,10,1000],'gamma':[0.001,0.000],'kernel':["rbf"]},

]

svc=svm.SVC(probability=True)
clf=GridSearchCV(svc,param_grid)
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
## Save the model using Pickle

import pickle

pickle.dump(clf,open('svm_model.pkl','wb'))
```

```python
model=pickle.load(open('svm_model.pkl','rb'))
```
